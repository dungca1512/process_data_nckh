```python
import matplotlib.pyplot as plt

def get_radar_chart(df):
    data = df.iloc[0]

    categories = list(data.index)
    N = len(categories)
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    data = list(data)
    data += data[:1]

    plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)
    plt.plot(angles, data, 'o-', linewidth=2)
    plt.fill(angles, data, alpha=0.25)
    plt.xticks(angles[:-1], categories)
    plt.yticks([0, 1, 2, 3, 4, 5], color='grey', size=8)
    plt.ylim(0, 6)
    plt.title('Radar chart', size=20)
    plt.show()
```

```python
get_radar_chart(df3)
```

```python
get_radar_chart(df4)
```
